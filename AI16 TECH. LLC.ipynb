```python
#import necessary libraries
import numpy as AI16N
import pandas as AI16P
from matplotlib import pyplot as AI16plt
from matplotlib import style
style.use("ggplot")
```

```python
#Data preparation
AI16 = AI16P.read_csv('desktop/AI16_TECH LLC.csv', sep=',', header = None)
AI16.columns=["QUALIFICATIONS","DISCIPLINE","AGE", "HONOURS", "DATA SCIENCE", "ARTIFICIAL INTELLIGENCE", "DEVELOPER", "LOCATION", "NYSC CERTIFICATE", "MARITAL STATUS", "WORK EXPERIENCE", "SALARY(NAIRA)"]
AI16.head()
```

```python
#Data analysis
#lets preview our dependent (x1-xn) and independent (y) variable
AI16X = AI16 [["QUALIFICATIONS","DISCIPLINE","AGE", "HONOURS", "DATA SCIENCE", "ARTIFICIAL INTELLIGENCE", "DEVELOPER", "LOCATION", "NYSC CERTIFICATE", "MARITAL STATUS", "WORK EXPERIENCE"]]
AI16Y = AI16 ["SALARY(NAIRA)"]
AI16X
AI16Y
```

```python
AI16X.shape
```

```python
AI16Y.shape
```

```python
#Data preprocessing
AI16drop_constant = AI16.drop(["DISCIPLINE" ,"LOCATION", "NYSC CERTIFICATE", "MARITAL STATUS"],  axis = "columns")
AI16drop_constant.tail()
```

```python
#one hot encoding
encoding = AI16P.get_dummies(AI16[["QUALIFICATIONS","HONOURS","DATA SCIENCE","ARTIFICIAL INTELLIGENCE","DEVELOPER"]])
encoding.head()
```

```python
#merge the encoded values to the data
AI16merged = AI16P.concat([AI16drop_constant, encoding], axis ="columns")
AI16merged
```

```python
AI16_finaldropX = AI16merged.drop(["QUALIFICATIONS","SALARY(NAIRA)" ,"HONOURS","DATA SCIENCE", "ARTIFICIAL INTELLIGENCE", "DEVELOPER"],  axis = "columns")
AI16_finaldropX.head()
```

```python
AI16Y.head()
```

```python
#linear regression algorithm
from sklearn.linear_model import LinearRegression
regr= LinearRegression()
from sklearn.model_selection import train_test_split
```

```python
#fit the model
X_train,X_test,Y_train,Y_test = train_test_split(AI16_finaldropX, AI16Y, test_size=0.25)
print(Y_train.shape, Y_test.shape)
```

```python
regr.fit(X_train,Y_train)
print("coefficients =" , regr.coef_)
print("intercepts =" , regr.intercept_)
```

```python
regr.predict(X_test)
```

```python
Y_test
```
